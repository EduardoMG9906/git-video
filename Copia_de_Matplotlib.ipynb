<h1><center><b>Matplotlib</b></center></h1>

<p>
Matplotlib es una librería de trazado 2D y 3D básico de Python. Su principal función es la generación de gráficos a partir de datos contenidos en lista o arreglos numpy en una gran variedad de formatos y entornos iteractivos como figuras o interfaces. Además de Python, Matplotlib puede ser usado también en servidores de aplicaciones web y librerías para la creación de interfaces gráficas de usuario (GUI por sus siglas en inglés). 
</p>
<p>
Las políticas de Matplotlib son hacer que las cosas fáciles sean fáciles y las difíciles sean posibles. Entre las múltiples cosas que se pueden hacer con esta librería encontramos diagramas, histogramas, espectros de potencia, gráficas de barras, visualización de imágenes, gráficas animadas entre otras. 
</p>
<p>
Matplotlib está soportado bajo una licencia de código abierto, por lo cual podemos adjuntar los gráficos generados en cualquier tipo de archivos sin consecuencias legales aparentes. Además cuenta con una comunidad en línea para el constante desarrollo de nuevas funcionalidades y depuración de errores. 
</p>
<p>
Esta librería viene incluida en el paquete de desarrollo cientifico <b>Scipy</b> junto con scipy y numpy.
</p>

<h3>Títulos, Etiquetas y Estilos</h3>

La información visualizada en un gráfico carecería de sentido sin un título o etiquetas que den ideas sobre que tipo de información es la presentada en este, cuales son sus medidas o magnitudes de las variables graficadas. Matplotlib soluciona este inconveniente de una manera sencilla, permitiendo agregar títulos, nombres de ejes y estilos con pocas líneas de código.

```python
#Usando los datos anteriores
plt.plot(x, y)
plt.title("Mi título")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

<p>Es posible dar estilo a los gráficos elaborados con matplotlib haciendo uso de los **kwargs admitidos como parámetros en la función <i>plot()</i>. Básicamente la estructura de la función sigue el siguiente modelo.</p>

<center>plt.plot(x, y, **kwarg)</center>

<p>El argumento **kwargs corresponde a un dicionario (clave, valor) el cual es procesado internamente por la función. Los parámetros admitidos pueden ser consultado en el siguiente enlace: <a href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html ">Click aquí</a></p>


```python
#Usando los datos anteriores
plt.figure(figsize=(15, 7))#Formatea el tamaño del gráfico
plt.plot(x, y, linestyle="--", color="red", marker="o")
plt.title("Mi título")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)#Agrega la cuadricula
plt.show()
```

<p>Los tres estilos vistos anteriormente dentro de la función <i>plot()</i> puden ser simplificados agregando una cadena de texto que codifique los estilos deseados siguiendo el siguiente orden: "estiloLinea+color+marcador", un ejemplo de esto es la cadena "--ro", en donde "--" representa el estilo de línea, "r" el color de la línea y "o" el marcador.</p>

```python
#Usando los datos anteriores
plt.figure(figsize=(15, 7))#Formatea el tamaño del gráfico
plt.plot(x, y, "--ro")
plt.title("Mi título")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)#Agrega la cuadricula
plt.show()
```

<p>Los estilos de líneas y marcadores permitidos por matplotlib pueden ser consultados en los siguientes enlaces: <a href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle">Estilos de línea</a> y <a href="https://matplotlib.org/3.1.1/api/markers_api.html">Marcadores</a></p>

<h3>Más sobre estilos</h3>

```python
#Se crean los datos a graficar usando Numpy
#Cremos el vector t.
t =  np.linspace(0, 10, 200)
a = 10 #Amplitud de la señal
f = 0.5 #Frecuencia de la señal
y = a*np.sin(2*np.pi*f*t)


plt.figure(figsize=(20,6))
plt.plot(t, y, "-gv", linewidth=2, markersize=8, markerfacecolor='orange', alpha=0.6)
plt.title('Gráfica seno')
plt.xlabel('Tiempo[ms]')
plt.ylabel('Amplitud[V]')
plt.grid(True)
plt.show()
```

<h1><center><b>Tipos De Gráficos</b></center></h1>

<p>Matplotlib ofrece en su distribución múltiples tipos de gráficos, el uso de ellos dependen de la los datos que se posean y la manera en personal en la cual se quieran mostrar dichos datos. A continuación de muestran algunos tipos de datos.</p>

<h3><b>Gráfico De Barras</b></h3>

Un gráfico de barras es una forma de resumir un conjunto de datos por categorías. Muestra los datos usando varias barras de la misma anchura, cada una de las cuales representa una categoría concreta.

```python
nombres = ['Enero','Febrero','Marzo','Abrir','Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
datos = [2,1,3,0,4,1,2,2,3,1,0,2]
x = range(len(datos))

plt.figure(figsize=(15,6)) # Figure
plt.bar(x, datos, width=0.8, color='orange')
#plt.xticks(x, nombres) #Más sobre estilos
plt.title("Cumpleaños por mes")
plt.ylabel("Número de personas")

plt.show()
```

<h3><b>Gráfico De Pastel</b></h3>
Sirve para representar variables cualitativas o discretas. Se utiliza para representar la proporción de elementos de cada uno de los valores de la variable.

```python
etiquetas = ['Perros', 'Gatos', 'Aves', 'Peces'] #Etiquetas que serán mostradas
porcentajes = [45, 30, 15, 10] #División en porcentajes.
explode = (0.05, 0, 0, 0) #Separa una muestra de los demás
autopct = '%1.1f%%' #Ayuda a mostrar los porcentajes dentro del diagrama

plt.figure(figsize=(6, 6))
plt.pie(porcentajes, explode=explode, labels=etiquetas, autopct=autopct,
        shadow=True, startangle=90)
plt.show()
```

<h3><b>Histograma</b></h3>

Muestra como es la distribución de un conjunto de datos.

```python
mu = 100 
sigma = 15 
x = mu + sigma * np.random.randn(10000)
print(x)

num_divisiones = 20
# the histogram of the data
n, bins, patches = plt.hist(x, num_divisiones, facecolor='blue')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
plt.show()
```

<h3><b>Gráficos En Escala Logarítmica</b></h3>

Lo que busca un gráfico en escala logarítmica es mantener constantes las proporciones entre los conjuntos de datos.

```python
# Data for plotting
t = np.arange(0.01, 20.0, 0.01)

# Eje Y en escala logarítmica
plt.semilogy(t, np.exp(-t / 5.0))
plt.title('semilogy')
plt.grid()
plt.show()

# Eje x en escala logaítmica
plt.semilogx(t, np.sin(2 * np.pi * t))
plt.title('semilogx')
plt.grid()
plt.show()

#Ambos ejes en escala logarítmica
plt.loglog(t, 20 * np.exp(-t / 10.0))
plt.title("loglog")
plt.grid()
plt.show()
```

<h3><b>Gráficos De Dispersión</b></h3>

Los gráficos de dispersión se usan para trazar puntos de datos en un eje vertical y uno horizontal, mediante lo que se trata de mostrar cuánto afecta una variable a otra.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()

colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

<h3><b>Gráficos Heatmap</b></h3>

```python
Z = np.random.rand(10, 10)
for i in range(0, 10):
  Z[i][i]=0

x = np.arange(0.5, 11, 1) 
y = np.arange(0.5, 11, 1) 
posiciones = np.arange(1, 11, 1)
canales = ["canal "+str(i) for i in range(1, 11)]

plt.figure(figsize=(10,8))
plt.pcolormesh(x, y, Z)
plt.colorbar()
plt.xticks(posiciones, canales)
plt.yticks(posiciones, canales)
plt.show()
```

<h3><b>Mostrando Imágenes Con Matplotlib</b></h3>

```python
import matplotlib.cbook as cbook
w, h = 512, 512

with cbook.get_sample_data('ct.raw.gz') as datafile:
    s = datafile.read()
A = np.frombuffer(s, np.uint16).astype(float).reshape((w, h))

plt.imshow(A)
plt.title('CT density')
plt.axis("off")

plt.show()
```

<h1><center><b>Graficación Múltiple</b></center></h1>

¿Es posible graficar distintos conjuntos de datos en una sola figura?

<h3>Misma figura, mismo axes<h3>

```python
#Creamos los datos a graficar
tiempo = np.linspace(0, 1, 100)
f = 3
seno = 4*np.sin(2*np.pi*tiempo*f)
coseno = 2*np.cos(2*np.pi*tiempo*f)

plt.plot(tiempo, seno, tiempo, coseno)
#plt.legend(["seno", "coseno"])
plt.show()
```

```python
#Se pueden graficar los mismos datos usando varias líneas de código
plt.figure(figsize=(10,6))
plt.plot(tiempo,seno,"-b", label="Seno")
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend(loc="best")
plt.show()
```

<h3>Misma figura, diferentes axes<h3>
Usando el metodo <i>subplot(n_filas, n_columnas, n)</i> es posible graficar los datos en la misma figura pero en diferentes axes. 

```python
#Usamos los datos anteriores
#Se configura para que se muestren en dos filas y una columna
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.grid()

plt.subplot(2,1,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.show()
```

```python
#Usamos los datos anteriores
#Se configura para que se muestren en dos filas y una columna
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.legend(loc="upper right")
plt.grid()

plt.subplot(1,2,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()

plt.subplot(2,2,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()

plt.subplot(2,2,3)
plt.imshow(A)
plt.title('CT density')
plt.axis("off")

plt.subplot(2,2,4)
plt.pie(porcentajes, explode=explode, labels=etiquetas, autopct=autopct,
        shadow=True, startangle=90)

plt.show()
```

<h1>Otra manera de generar figuras</h1>

```python
fig, axes = plt.subplots()
axes.plot(tiempo, coseno, ".-r",label="Coseno")
axes.grid()
axes.legend()
plt.show()
```

```python
fig, (axes1, axes2) = plt.subplots(1,2)
axes1.pie(porcentajes, explode=explode, labels=etiquetas, autopct=autopct,
        shadow=True, startangle=90)
axes2.plot(tiempo,seno,"-b", label="Seno")
axes2.plot(tiempo, coseno, ".-r",label="Coseno")
axes2.legend()
axes2.grid()
plt.show()
```
