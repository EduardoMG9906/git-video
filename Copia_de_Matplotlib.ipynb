<h3>Ttulos, Etiquetas y Estilos</h3>

La informacin visualizada en un grfico carecera de sentido sin un ttulo o etiquetas que den ideas sobre que tipo de informacin es la presentada en este, cuales son sus medidas o magnitudes de las variables graficadas. Matplotlib soluciona este inconveniente de una manera sencilla, permitiendo agregar ttulos, nombres de ejes y estilos con pocas lneas de cdigo.

```python
#Usando los datos anteriores
plt.plot(x, y)
plt.title("Mi ttulo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

<p>Es posible dar estilo a los grficos elaborados con matplotlib haciendo uso de los **kwargs admitidos como parmetros en la funcin <i>plot()</i>. Bsicamente la estructura de la funcin sigue el siguiente modelo.</p>

<center>plt.plot(x, y, **kwarg)</center>

<p>El argumento **kwargs corresponde a un dicionario (clave, valor) el cual es procesado internamente por la funcin. Los parmetros admitidos pueden ser consultado en el siguiente enlace: <a href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html ">Click aqu</a></p>


```python
#Usando los datos anteriores
plt.figure(figsize=(15, 7))#Formatea el tamao del grfico
plt.plot(x, y, linestyle="--", color="red", marker="o")
plt.title("Mi ttulo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)#Agrega la cuadricula
plt.show()
```

<p>Los tres estilos vistos anteriormente dentro de la funcin <i>plot()</i> puden ser simplificados agregando una cadena de texto que codifique los estilos deseados siguiendo el siguiente orden: "estiloLinea+color+marcador", un ejemplo de esto es la cadena "--ro", en donde "--" representa el estilo de lnea, "r" el color de la lnea y "o" el marcador.</p>

```python
#Usando los datos anteriores
plt.figure(figsize=(15, 7))#Formatea el tamao del grfico
plt.plot(x, y, "--ro")
plt.title("Mi ttulo")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(True)#Agrega la cuadricula
plt.show()
```

<p>Los estilos de lneas y marcadores permitidos por matplotlib pueden ser consultados en los siguientes enlaces: <a href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle">Estilos de lnea</a> y <a href="https://matplotlib.org/3.1.1/api/markers_api.html">Marcadores</a></p>

<p>Matplotlib ofrece en su distribucin mltiples tipos de grficos, el uso de ellos dependen de la los datos que se posean y la manera en personal en la cual se quieran mostrar dichos datos. A continuacin de muestran algunos tipos de datos.</p>

<h3><b>Grfico De Barras</b></h3>

Un grfico de barras es una forma de resumir un conjunto de datos por categoras. Muestra los datos usando varias barras de la misma anchura, cada una de las cuales representa una categora concreta.

```python
nombres = ['Enero','Febrero','Marzo','Abrir','Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
datos = [2,1,3,0,4,1,2,2,3,1,0,2]
x = range(len(datos))

plt.figure(figsize=(15,6)) # Figure
plt.bar(x, datos, width=0.8, color='orange')
#plt.xticks(x, nombres) #Ms sobre estilos
plt.title("Cumpleaos por mes")
plt.ylabel("Nmero de personas")

plt.show()
```

<h3><b>Grfico De Pastel</b></h3>
Sirve para representar variables cualitativas o discretas. Se utiliza para representar la proporcin de elementos de cada uno de los valores de la variable.

```python
etiquetas = ['Perros', 'Gatos', 'Aves', 'Peces'] #Etiquetas que sern mostradas
porcentajes = [45, 30, 15, 10] #Divisin en porcentajes.
explode = (0.05, 0, 0, 0) #Separa una muestra de los dems
autopct = '%1.1f%%' #Ayuda a mostrar los porcentajes dentro del diagrama

plt.figure(figsize=(6, 6))
plt.pie(porcentajes, explode=explode, labels=etiquetas, autopct=autopct,
        shadow=True, startangle=90)
plt.show()
```

<h3><b>Histograma</b></h3>

Muestra como es la distribucin de un conjunto de datos.

```python
mu = 100 
sigma = 15 
x = mu + sigma * np.random.randn(10000)
print(x)

num_divisiones = 20
# the histogram of the data
n, bins, patches = plt.hist(x, num_divisiones, facecolor='blue')

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')
plt.show()
```

<h3><b>Grficos En Escala Logartmica</b></h3>

Lo que busca un grfico en escala logartmica es mantener constantes las proporciones entre los conjuntos de datos.

```python
# Data for plotting
t = np.arange(0.01, 20.0, 0.01)

# Eje Y en escala logartmica
plt.semilogy(t, np.exp(-t / 5.0))
plt.title('semilogy')
plt.grid()
plt.show()

# Eje x en escala logatmica
plt.semilogx(t, np.sin(2 * np.pi * t))
plt.title('semilogx')
plt.grid()
plt.show()

#Ambos ejes en escala logartmica
plt.loglog(t, 20 * np.exp(-t / 10.0))
plt.title("loglog")
plt.grid()
plt.show()
```

<h3><b>Grficos De Dispersin</b></h3>

Los grficos de dispersin se usan para trazar puntos de datos en un eje vertical y uno horizontal, mediante lo que se trata de mostrar cunto afecta una variable a otra.

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()

colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

<h3><b>Grficos Heatmap</b></h3>

```python
Z = np.random.rand(10, 10)
for i in range(0, 10):
  Z[i][i]=0

x = np.arange(0.5, 11, 1) 
y = np.arange(0.5, 11, 1) 
posiciones = np.arange(1, 11, 1)
canales = ["canal "+str(i) for i in range(1, 11)]

plt.figure(figsize=(10,8))
plt.pcolormesh(x, y, Z)
plt.colorbar()
plt.xticks(posiciones, canales)
plt.yticks(posiciones, canales)
plt.show()
```

<h3><b>Mostrando Imgenes Con Matplotlib</b></h3>

```python
import matplotlib.cbook as cbook
w, h = 512, 512

with cbook.get_sample_data('ct.raw.gz') as datafile:
    s = datafile.read()
A = np.frombuffer(s, np.uint16).astype(float).reshape((w, h))

plt.imshow(A)
plt.title('CT density')
plt.axis("off")

plt.show()
```

<h1><center><b>Graficacin Mltiple</b></center></h1>

Es posible graficar distintos conjuntos de datos en una sola figura?

<h3>Misma figura, mismo axes<h3>

```python
#Creamos los datos a graficar
tiempo = np.linspace(0, 1, 100)
f = 3
seno = 4*np.sin(2*np.pi*tiempo*f)
coseno = 2*np.cos(2*np.pi*tiempo*f)

plt.plot(tiempo, seno, tiempo, coseno)
#plt.legend(["seno", "coseno"])
plt.show()
```

```python
#Se pueden graficar los mismos datos usando varias lneas de cdigo
plt.figure(figsize=(10,6))
plt.plot(tiempo,seno,"-b", label="Seno")
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend(loc="best")
plt.show()
```

<h3>Misma figura, diferentes axes<h3>
Usando el metodo <i>subplot(n_filas, n_columnas, n)</i> es posible graficar los datos en la misma figura pero en diferentes axes. 

```python
#Usamos los datos anteriores
#Se configura para que se muestren en dos filas y una columna
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.grid()

plt.subplot(2,1,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.show()
```

```python
#Usamos los datos anteriores
#Se configura para que se muestren en dos filas y una columna
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.legend(loc="upper right")
plt.grid()

plt.subplot(1,2,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(tiempo,seno,"-b", label="Seno")
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()

plt.subplot(2,2,2)
plt.plot(tiempo, coseno, ".-r",label="Coseno")
plt.legend()
plt.grid()

plt.subplot(2,2,3)
plt.imshow(A)
plt.title('CT density')
plt.axis("off")

plt.subplot(2,2,4)
plt.pie(porcentajes, explode=explode, labels=etiquetas, autopct=autopct,
        shadow=True, startangle=90)

plt.show()
```
